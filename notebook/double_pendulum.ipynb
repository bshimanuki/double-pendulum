\begin{align}
	M &= \begin{bmatrix}
		I_1+I_2+m_2l_1^2+2m_2l_1l_{c2}\cos\theta_2 & I_2+m_2l_1l_{c2}\cos\theta_2 \\
		I_2+m_2l_1l_{c2}\cos\theta_2 & I_2
	\end{bmatrix} \\
    M^{-1} &= \frac1{I_1I_2+m_2I_1^2I_2-m_2^2l_1^2l_{c2}^2\cos^2\theta_2} \begin{bmatrix}
		I_2 & -I_2-m_2l_1l_{c2}\cos\theta_2 \\
		-I_2-m_2l_1l_{c2}\cos\theta_2 & I_1+I_2+2m_2l_1l_{c2}\cos\theta_2
	\end{bmatrix} \\
	C &= \begin{bmatrix}
		-2m_2l_1l_{c2}\dot\theta_2\sin\theta_2 & -m_2l_1l_{c2}\dot\theta_2\sin\theta_2 \\
		m_2l_1l_{c2}\dot\theta_1\sin\theta_2 & 0
	\end{bmatrix} \\
	\tau_g &= \begin{bmatrix}
		-m_1gl_{c1}\sin\theta_1-m_2g(l_1\sin\theta_1+l_{c2}\sin(\theta_1+\theta_2)) \\
		-m_2gl_{c2}\sin(\theta_1+\theta_2)
	\end{bmatrix} \\
	B &= \begin{bmatrix}
		1 \\ 0
	\end{bmatrix} \\
	x &= \begin{bmatrix}
		\theta_1 \\ \theta_2 \\ \dot \theta_1 \\ \dot \theta_2
	\end{bmatrix} \\
	M\ddot q+C\dot q &= \tau_g+Bu \\
	\ddot q &= M^{-1}(\tau_g + Bu - C\dot q) \\
\end{align}

\begin{align}
	M^{-1} &= \frac1{I_1I_2+m_2I_1^2I_2-m_2^2l_1^2l_{c2}^2\cos^2\theta_2} \begin{bmatrix}
		I_2 & -I_2-m_2l_1l_{c2}\cos\theta_2 \\
		-I_2-m_2l_1l_{c2}\cos\theta_2 & I_1+I_2+2m_2l_1l_{c2}\cos\theta_2
	\end{bmatrix} \\
	M^{-1}_{q_1=\pi,q_2=0} &= \frac1{I_1I_2+m_2I_1^2I_2-m_2^2l_1^2l_{c2}^2} \begin{bmatrix}
		I_2 & -I_2+m_2l_1l_{c2} \\
		-I_2+m_2l_1l_{c2} & I_1+I_2-2m_2l_1l_{c2}
	\end{bmatrix} \\
	\frac{\partial\tau_g}{\partial q} &= \begin{bmatrix}
		-m_1gl_{c1}\cos\theta_1 - m_2g(l_1\cos\theta_1+l_{c2}\cos(\theta_1+\theta_2)) & -m_2gl_{c2}\cos(\theta_1+\theta_2) \\
		-m_2gl_{c2}\cos(\theta_1+\theta_2) & -m_2gl_{c2}\cos(\theta_1+\theta_2) \\
	\end{bmatrix} \\
	\frac{\partial\tau_g}{\partial q}_{q_1=\pi,q_2=0} &= \begin{bmatrix}
		m_1gl_{c1} + m_2g(l_1+l_{c2}) & m_2gl_{c2} \\
		m_2gl_{c2} & m_2gl_{c2} \\
	\end{bmatrix} \\
	\frac{\partial\tau_g}{\partial q}_{q_1=\pi,q_2=0} &= \begin{bmatrix}
		m_1gl_{c1} + m_2gl_1 & 0 \\
		0 & 0 \\
	\end{bmatrix} + m_2gl_{c2} \\
	\dot x &= \begin{bmatrix}
		\dot q \\
		M^{-1}(\tau_g+Bu-C\dot q)
	\end{bmatrix} \\
	\dot x &\approx A_{lin}(x-x^*) + B_{lin}(u-u^*) \\
	A_{lin} &= \begin{bmatrix}
		O_2 & I_2 \\
		M^{-1}(\frac{\partial\tau_g}{\partial q} + \frac{\partial B}{\partial q}u) & O_2
	\end{bmatrix}_{x=x^*,u=u^*} \\
	A_{lin} &= \begin{bmatrix}
		O_2 & I_2 \\
		M^{-1}\frac{\partial\tau_g}{\partial q} & O_2
	\end{bmatrix}_{x=x^*,u=u^*} \\
	B_{lin} &= \begin{bmatrix}
		\vec0_2 \\ M^{-1}B
	\end{bmatrix}_{x=x^*,u=u^*} \\
	A_{lin} &= \begin{bmatrix}
		O_2 & I_2 \\
		M^{-1}\frac{\partial\tau_g}{\partial q} & O_2
	\end{bmatrix}_{x=x^*,u=u^*} \\
	\dot x &= \begin{bmatrix}
		I_2 & O_2 \\
		O_2 & M^{-1}
	\end{bmatrix} \left(\begin{bmatrix}
			O_2 & I_2 \\
			O_2 & -C
		\end{bmatrix} x + \begin{bmatrix}
			\vec0_2 \\ \tau_g
		\end{bmatrix} + \begin{bmatrix}
			\vec0_2 \\ B
	\end{bmatrix} u\right) \\
	\dot x &= \begin{bmatrix}
		I_2 & O_2 \\
		O_2 & M^{-1}
	\end{bmatrix} \left(\begin{bmatrix}
			O_2 & I_2 \\
			O_2 & O_2
		\end{bmatrix} x + \begin{bmatrix}
			0 \\ 0 \\ -(m_1 l_1 + m_2 l_2) g sin(\theta_1) \\ 0
		\end{bmatrix} + \begin{bmatrix}
			0 \\ 0 \\ 1 \\ 0
	\end{bmatrix} u\right) \\
	\dot x &\approx \begin{bmatrix}
		I_2 & O_2 \\
		O_2 & M^{-1}
	\end{bmatrix} \left(\begin{bmatrix}
			O_2 & I_2 \\
			O_2 & O_2
		\end{bmatrix} x + \begin{bmatrix}
			0 \\ 0 \\ -(m_1 l_1 + m_2 l_2) g (\sin(\theta_{1f})+\cos(\theta_{1f})(\theta_1-\theta_{1f})) \\ 0
		\end{bmatrix} + \begin{bmatrix}
			0 \\ 0 \\ 0 \\ 1
	\end{bmatrix} u\right) \\
	\dot x &\approx \begin{bmatrix}
		I_2 & O_2 \\
		O_2 & M^{-1}
	\end{bmatrix} \left(\begin{bmatrix}
			0 & 0 & 1 & 0 \\

			0 \\ 0 \\ 0 \\ 1

\end{align}

```python
import matplotlib.pyplot as plt
import time

controller = double_pendulum.make_controller()

# Calculates the closed loop f(x) at xn
def calcF(xn):
    '''
    Code submission for 3.5: populate
    this function to calculate the closed-loop
    system dynamics f(x) at the input point.
    '''
    u = controller.lqr_controller(xn)
    return controller.evaluate_f(xn, u)

# Calculates V(xn)
def calcV(xn):
    '''
    Code submission for 3.5: populate
    this function to calculate V(x)
    at the input point.
    '''
    xn = np.array(xn)
    xn -= xf
    xn[:2] = np.unwrap(xn[:2])
    S = controller.S
    v = np.dot(xn.T, np.dot(S, xn)).item()
    return v

# Calculates \dot{V}(xn).
def calcVdot(xn):
    '''
    Code submission for 3.5: populate
    this function to calculate Vdot(x)
    at the input point.
    '''
    xn = np.array(xn)
    xnbar = xn - xf
    xnbar[:2] = np.unwrap(xnbar[:2])
    S = controller.S
    vdot = (np.dot(xnbar.T, np.dot(S, calcF(xn))) + np.dot(calcF(xn).T, np.dot(S, xnbar))).item()
    return vdot

start_time = time.time()

# Sample f, V, and Vdot over
# a grid defined by these parameters.
# (Odd numbers are good because there'll be
# a bin at exactly the origin.
# These are slightly strange numbers as we've
# tried to default these to something as small
# as possible while still giving reasonable results.
# Feel free to increase if your computer and patience
# can handle it.)
n_bins = 21
n_bins_thetad = 7
theta1_width = 1
theta2_width = 1
theta1d_width = 10
theta2d_width = 20

# Do the actual sampling....
x = np.linspace(xf[0]-theta1_width, xf[0]+theta1_width, n_bins)
y = np.linspace(xf[1]-theta2_width, xf[1]+theta2_width, n_bins)
z = np.linspace(xf[2]-theta1d_width, xf[2]+theta1d_width, n_bins_thetad)
w = np.linspace(xf[3]-theta2d_width, xf[3]+theta2d_width, n_bins_thetad)
X, Y, Z, W = np.meshgrid(x, y, z, w, indexing="ij")



def calc_over_array(f):
    return np.array([[[[f([dx, dy, dz, dw]) for dw in w] for dz in z] for dx in x] for dy in y])

V_samples = calc_over_array(calcV)

f_samples = calc_over_array(calcF)
Vdot_samples = calc_over_array(calcVdot)

#xn = [np.pi-0,0,-0.01,-0.01]
#xnbar = xn - xf
#print(np.dot(xnbar.T, np.dot(controller.S, calcF(xn))) + np.dot(calcF(xn).T, np.dot(controller.S, xnbar)))
#print(calcVdot(xn))

elapsed = time.time() - start_time
print "Computed %d x %d x %d x %d sampling in %f seconds" % (n_bins, n_bins, n_bins_thetad, n_bins_thetad, elapsed)

#print "rho = ", estimate_rho(V_samples, Vdot_samples)
```

```python
# This cell plots the samples using color-coded plots.
# Color coding:
#   V: blue = low-value, red = high-value
#   Vdot: blue = low value, yellow = around 0, red = high value
# The plot of Vdot is overlayed with a quiver plot of the samples
# of f.

# Select with slice of \dot{theta_2} we'll
# plot... this slice should be close to 0,
# as it's the middle bin.
theta1d_plotting_slice = n_bins_thetad / 2
theta2d_plotting_slice = n_bins_thetad / 2

plt.figure().set_size_inches(12,12)

# Plot V
Xplot, Yplot = np.meshgrid(x, y)
plt.subplot(2, 1, 1)
plt.pcolormesh(Xplot, Yplot, V_samples[:, :, theta1d_plotting_slice, theta2d_plotting_slice])
plt.title("V(x) at dtheta_1 = %f, dtheta_2 = %f" % (z[theta1d_plotting_slice], w[theta2d_plotting_slice]))
plt.xlabel("theta_1")
plt.ylabel("theta_2")
plt.colorbar()

# Plot Vdot
# Use a sigmoid to try to squash the huge range of Vdot
# into something more visually appealing.
def sigmoid(x):
    return 1 / (1 + np.exp(-x/1e2))

plt.subplot(2, 1, 2)
Vdot_viz = sigmoid(Vdot_samples[:, :, theta1d_plotting_slice, theta2d_plotting_slice])
plt.pcolormesh(Xplot, Yplot, Vdot_viz, vmin=0, vmax=1.0)
plt.title("Vdot(x) overlaid with phase diagram f(x) at dtheta_1 = %f, dtheta_2 = %f" % (z[theta1d_plotting_slice], w[theta2d_plotting_slice]))
plt.xlabel("theta_1")
plt.ylabel("theta_2")

# Don't plot a quiver arrow at *every* point, but instead
# every ds points
# (lower = more quiver arrows)
ds = 3
plt.quiver(Xplot[::ds, ::ds], 
           Yplot[::ds, ::ds], 
           f_samples[::ds, ::ds, theta1d_plotting_slice, theta2d_plotting_slice, 2],
           f_samples[::ds, ::ds, theta1d_plotting_slice, theta2d_plotting_slice, 3]);
```

```python
def estimate_rho(V, Vdot):
    '''
    Code submission for 3.6
    Fill in this function to use the samples of V and Vdot
    (Each array has dimension [n_bins, n_bins, n_bins_theta2d])
    to compute a maximal rho indicating the region of attraction
    of the fixed point at the upright.
    '''
    rho_plus_epsilon = np.min(np.where((Vdot >= 0) & (V > 0), V, np.full_like(V, np.inf)))
    rho = np.max(np.where(V < rho_plus_epsilon, V, np.zeros_like(V)))
    return rho

print "V(x) = ", calcV(np.array([np.pi, 0, 0.1, 0.1]))

rho = estimate_rho(V_samples, Vdot_samples)
print "Region of attraction estimated at V(x) <= ", rho

# Plot Vdot again, but overlay the region of attraction -- which,
# for quadratic V, is an ellipse.
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
fig.set_size_inches(12,6)
plt.pcolormesh(Xplot, Yplot, Vdot_viz, vmin=0, vmax=1.0)

# The part of S we care about is the 2x2 submatrix from the 1st and 2nd rows
# and columns.
S_sub = controller.S[np.ix_([0, 1], [0, 1])]
# Extract its eigenvalues and eigenvectors, which tell us
# the axes of the ellipse
ellipseInfo = np.linalg.eig(S_sub)
# Eigenvalues are 1/r^2, Eigenvectors are axis directions
axis_1 = ellipseInfo[1][0, :]
if ellipseInfo[0][0] > 0 and ellipseInfo[0][1] > 0:
    r1 = math.sqrt(rho)/math.sqrt(ellipseInfo[0][0])
    axis_2 = ellipseInfo[1][1, :]
    r2 = math.sqrt(rho)/math.sqrt(ellipseInfo[0][1])
    angle = math.atan2(-axis_1[1], axis_1[0])
    from matplotlib.patches import Ellipse
    ax.add_patch(Ellipse((xf[0], xf[1]), 
                         2*r1, 2*r2, 
                         angle=angle*180./math.pi, 
                         linewidth=2, fill=False, zorder=2));
    
    # Report an interesting number that is easy to compute
    # from the ellipse info
    print "Area of your region of attraction: ", math.pi * r1 * r2
else:
    print "S_sub had nonpositive eigenvalues. That shouldn't happen."
    
plt.title("Vdot(x) overlaid with estimated ROA at dtheta_1 = %f, dtheta_2 = %f" % (z[theta1d_plotting_slice], w[theta2d_plotting_slice]))
plt.xlabel("theta_1")
plt.ylabel("theta_2");
```
